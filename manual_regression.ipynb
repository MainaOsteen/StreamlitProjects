```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    {
        "study_hours": [2, 3, 4],
        "scores": [83, 85, 87]

    }
)

df
```

```python
def regression_line(x, y):

   n = len(x)

   sum_xy = sum(a*b for a, b in zip(x, y))
   sum_x = sum(x)
   sum_y = sum(y)

   sum_x2 = sum(k**2 for k in x)
   sum_x_sqr = sum_x ** 2

   b = ((n * sum_xy) - (sum_x * sum_y)) / ((n * sum_x2) - sum_x_sqr)

   a = ((sum_x2 * sum_y) - (sum_x * sum_xy)) / ((n * sum_x2) - sum_x_sqr)

   return round(a, 4), round(b, 4)
```

```python
slope, intercept  = regression_line(df["study_hours"], df["scores"])

print(f"The equation of the line is: {slope:.4f}x + {intercept:.4f}")
```
