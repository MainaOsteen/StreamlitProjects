```python
x = [11, 13, 15, 17, 23 ,27, 31]
y = [120, 135, 145, 161, 170, 188, 194]
n = len(x)

sum_x = sum(x)
sum_y = sum(y)
sum_xy = sum(val_x*val_y for val_x, val_y in zip(x, y))
sum_x_sq = sum(val_x**2 for val_x in x)

slope = (n * sum_xy - sum_x + sum_y) / (n * sum_x_sq - sum_x**2)

intercept = (sum_y - slope * sum_x) / n

print(f"Manual Result: y  = {slope:.2f}x + {intercept:.3f}")
```
