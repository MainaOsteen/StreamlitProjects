```python
original = "123456" 


def fake_bin(x):

    final_string = ""

    
    for digit in x:

        num = int(digit)

        if(num < 5):
            final_string += "0"
        
        else:
            final_string += "1"

    return final_string
    


replaced = fake_bin(original)
print(replaced)

        
```
